## 💡 Tips

- **Filtre** en list comprehension : `[x for x in lst if condition]`
- **Transformation** : `[f(x) for x in lst]`
- **Ternaire** : `[a if condition else b for x in lst]`
- Une fonction doit opérer sur ses **paramètres**, jamais sur des variables globales
- `range(start, stop, step)` — `stop` est **exclu**
- Pour accumuler dans une boucle : initialise ta variable **avant** la boucle (`total = 0`, `result = []`)
- `enumerate(lst)` donne `(index, valeur)` à chaque itération — très utile

## Partie 1 — Fonctions

### 1.1 — Classifier un IMC

Écris une fonction `classifier_imc(imc)` qui retourne :
- `"Insuffisance pondérale"` si imc < 18.5
- `"Poids normal"` si 18.5 <= imc < 25
- `"Surpoids"` si 25 <= imc < 30
- `"Obésité"` si imc >= 30

Teste avec les valeurs : `16.0`, `22.5`, `27.3`, `31.8`


````python
# Ton code ici
def classifier_imc(imc):
    if imc < 18.5:
        return "Insuffisance pondérale"
    elif 18.5 <= imc < 25:
        return "Poids normal"
    elif 25 <= imc < 30:
        return "Surpoids"
    elif imc >= 30:
        return "Obésité"

print("Case_1:", classifier_imc(16.0))
print("Case_2:", classifier_imc(22.5))
print("Case_3:", classifier_imc(27.3))
print("Case_4:", classifier_imc(31.8))
````

### 1.2 — Fonction avec accumulation

Écris une fonction `compter_recus(scores, seuil=60)` qui :
- Prend une liste de scores et un seuil
- Retourne le **nombre** de scores >= seuil (avec une boucle `for`, pas `sum()` ni list comprehension)

Teste avec `scores = [45, 72, 58, 88, 61, 39, 95, 50, 67]` et :
1. Le seuil par défaut
2. `seuil=70`
3. `seuil=90`


````python
def compter_recus(scores, seuil=60):
    score_list = []
    for score in scores:
        if score >= seuil:
            score_list.append(score)
    score_nb = len(score_list)
    return score_nb

print("Seuil par défault:", compter_recus([45, 72, 58, 88, 61, 39, 95, 50, 67]))
print("Seuil 70:", compter_recus([45, 72, 58, 88, 61, 39, 95, 50, 67], 70))
print("Seuil 90:", compter_recus([45, 72, 58, 88, 61, 39, 95, 50, 67], 90))
````

## Partie 2 — List Comprehensions

### 2.1 — Transformation + filtre enchaînés

Tu as cette liste de températures en Celsius :
```python
temperatures_c = [0, 15, -5, 37, 22, 100, -10, 28]
```

1. Crée `temperatures_f` : conversion en Fahrenheit pour **toutes** les températures (`F = C * 9/5 + 32`)
2. Crée `temperatures_positives` : uniquement les températures Celsius **strictement positives**
3. Crée `alertes` : liste de chaînes — `"Chaud"` si C > 30, `"Froid"` si C < 0, `"Normal"` sinon

Affiche les 3 listes.


````python
temperatures_c = [0, 15, -5, 37, 22, 100, -10, 28]

temperatures_f = [(temperature * 9/5 + 32) for temperature in temperatures_c]
print("temperature Fahrenheit = ",temperatures_f)

temperatures_positives = [temperature for temperature in temperatures_c if temperature > 0]
print("Positive temperatures = ", temperatures_positives)

alertes = ["Chaud" if temperature > 30 else "Froid" if temperature < 0 else "Normal" for temperature in temperatures_c]
print("Alertes:", alertes)
````

## Partie 3 — Boucles

### 3.1 — enumerate()

Tu as cette liste de candidats :
```python
candidats = ["Alice", "Bob", "Clara", "David", "Eva"]
```

Affiche chaque candidat avec son **rang** (en commençant à 1, pas à 0) :
```
1. Alice
2. Bob
...
```

Utilise `enumerate()`.


````python
candidats = ["Alice", "Bob", "Clara", "David", "Eva"]

for rang, candidat in enumerate(candidats, start=1):
    print(f"{rang}, {candidat}")
````

### 3.2 — Boucle + condition + accumulation

```python
transactions = [120, -45, 200, -30, 80, -15, 350, -200, 60]
```

Les valeurs positives sont des **crédits**, les négatives des **débits**.

Avec une seule boucle `for` :
1. Calcule le total des crédits
2. Calcule le total des débits (en valeur absolue)
3. Calcule le solde final (crédits + débits, les débits étant négatifs)

Affiche les 3 résultats.


````python
transactions = [120, -45, 200, -30, 80, -15, 350, -200, 60]

credits = 0
debits = 0
for transaction in transactions:
    if transaction > 0:
        credits += transaction
    elif transaction < 0:
        debits += transaction
print("Crédit = ", credits)
print("Débit = ", abs(debits))
print("Solde final = ", credits + debits, "euros")
````

### 3.3 — Boucle imbriquée

Tu as deux listes :
```python
equipes = ["Alpha", "Beta", "Gamma"]
membres = ["Alice", "Bob", "Clara"]
```

Génère et affiche toutes les **combinaisons possibles** (chaque membre dans chaque équipe) :
```
Alice — Alpha
Alice — Beta
...
```

Puis compte et affiche le nombre total de combinaisons.


````python
equipes = ["Alpha", "Beta", "Gamma"]
membres = ["Alice", "Bob", "Clara"]

for membre in membres:
    for equipe in equipes:
        print(f"{equipe} - {membre}")

print("Number of combinations = ", (len(equipes) * len(membres)))
````

## 🎯 Bonus — Tout combiner

```python
ventes = [
    {"produit": "Laptop", "categorie": "Tech", "quantite": 3, "prix_unitaire": 1200},
    {"produit": "Chaise", "categorie": "Mobilier", "quantite": 10, "prix_unitaire": 150},
    {"produit": "Ecran", "categorie": "Tech", "quantite": 5, "prix_unitaire": 400},
    {"produit": "Bureau", "categorie": "Mobilier", "quantite": 2, "prix_unitaire": 600},
    {"produit": "Clavier", "categorie": "Tech", "quantite": 8, "prix_unitaire": 80},
    {"produit": "Lampe", "categorie": "Mobilier", "quantite": 15, "prix_unitaire": 45},
]
```

1. Crée une list comprehension `ventes_tech` : uniquement les ventes de catégorie `"Tech"`
2. Écris une fonction `chiffre_affaires(liste_ventes)` qui retourne le CA total (`quantite * prix_unitaire` pour chaque vente, sommé). Teste-la sur `ventes` et sur `ventes_tech`.
3. Avec une boucle, affiche chaque produit et son CA individuel, trié par ordre décroissant de CA.

> 💡 Pour le tri : tu peux construire une liste de tuples `(produit, ca)` puis utiliser `sorted(..., key=lambda x: x[1], reverse=True)`


````python
ventes = [
    {"produit": "Laptop", "categorie": "Tech", "quantite": 3, "prix_unitaire": 1200},
    {"produit": "Chaise", "categorie": "Mobilier", "quantite": 10, "prix_unitaire": 150},
    {"produit": "Ecran", "categorie": "Tech", "quantite": 5, "prix_unitaire": 400},
    {"produit": "Bureau", "categorie": "Mobilier", "quantite": 2, "prix_unitaire": 600},
    {"produit": "Clavier", "categorie": "Tech", "quantite": 8, "prix_unitaire": 80},
    {"produit": "Lampe", "categorie": "Mobilier", "quantite": 15, "prix_unitaire": 45},
]

ventes_tech = [vente for vente in ventes if vente['categorie'] == "Tech"]
print("Ventes dans la tech:", ventes_tech)
````

````python
def chiffre_affaires(liste_ventes):
    ca_total = sum([vente["quantite"] * vente["prix_unitaire"] for vente in liste_ventes])
    return ca_total

print("CA ventes = ", chiffre_affaires(ventes), "euros")
print("CA ventes_tech = ", chiffre_affaires(ventes_tech), "euros")
````

````python
ca_produits = []

for vente in ventes:
    ca = vente["quantite"] * vente["prix_unitaire"]
    ca_produits.append((vente["produit"], ca))

ca_produits = sorted(ca_produits, key=lambda x: x[1], reverse=True)

for produit, ca in ca_produits:
    print(f"{produit} : {ca} €")
````
