## 💡 Tips

- `str.split(sep)` découpe une chaîne → retourne une liste. Ex : `"a,b,c".split(",")` → `["a", "b", "c"]`
- `sep.join(liste)` fait l'inverse → `",".join(["a", "b", "c"])` → `"a,b,c"`
- `str.strip()` supprime les espaces (et `\n`) en début/fin
- `dict.get(clé, valeur_par_défaut)` — évite un `KeyError` si la clé n'existe pas
- Pour compter des occurrences : initialise un dict vide, puis `dict[clé] = dict.get(clé, 0) + 1`
- Une fonction peut appeler une autre fonction — c'est même recommandé pour découper la logique
- Boucle `while` : vérifie une condition **avant** chaque itération. Toujours prévoir une condition de sortie pour éviter une boucle infinie

## Partie 1 — Manipulation de strings

### 1.1 — Parser une chaîne

Tu reçois des données brutes sous forme de string :
```python
ligne = "  Alice,Data Scientist,52000,Paris  "
```

1. Nettoie la chaîne (supprime les espaces en début/fin)
2. Découpe-la en une liste de 4 éléments → `champs`
3. Affiche chaque champ avec son index (utilise `enumerate`)
4. Crée un dict `employe` avec les clés `"nom"`, `"poste"`, `"salaire"` (converti en `int`), `"ville"`


````python
ligne = "  Alice,Data Scientist,52000,Paris  "

ligne_cleaned = ligne.strip()
print("ligne_cleaned:", ligne_cleaned)

champs = ligne_cleaned.split(",")
print(champs)

for rang, champ in enumerate(champs, start=1):
    print(f"{rang} - {champ}")

employe = {
    "nom": "Alice",
    "poste": "Data Scientist",
    "salaire": int(52000),
    "ville": "Paris"
}
print(employe)
````

## Partie 2 — Dictionnaires avancés

### 2.1 — Compter des occurrences

Tu as cette liste de logs d'événements :
```python
logs = ["login", "logout", "login", "error", "login", "error", "logout", "login", "error", "error"]
```

Sans utiliser `Counter` de `collections` :
1. Crée un dict `comptage` qui associe chaque événement à son nombre d'occurrences
2. Affiche les événements triés par fréquence décroissante (utilise `sorted` + `lambda`)

> 💡 `dict.get(clé, 0)` retourne `0` si la clé n'existe pas encore


````python
logs = ["login", "logout", "login", "error", "login", "error", "logout", "login", "error", "error"]

comptage={"login": 0,
          "logout": 0,
          "error": 0}

for log in logs:
    if log == "login":
        comptage["login"] += 1
    elif log == "logout":
        comptage["logout"] += 1
    elif log == "error":
        comptage["error"] += 1
print("comptage:", comptage)

comptage = sorted(comptage.items(), key=lambda x: x[1], reverse=True)
print("Comptage after sorting:", comptage)
````

### 2.2 — Construire un dict depuis une boucle

Tu as deux listes :
```python
noms = ["Alice", "Bob", "Clara", "David"]
scores = [88, 72, 95, 61]
```

1. Crée un dict `resultats` qui associe chaque nom à son score — avec une boucle `for` (pas `zip()` ni dict comprehension)
2. Ajoute une clé `"moyenne"` dans `resultats` avec la moyenne des scores (arrondie à 1 décimale)
3. Affiche uniquement les candidats avec un score **supérieur à la moyenne**

> 💡 Pour itérer sur deux listes en parallèle sans `zip()` : utilise `range(len(noms))` et indexe les deux listes


````python
noms = ["Alice", "Bob", "Clara", "David"]
scores = [88, 72, 95, 61]

resultats={}
for i in range(len(noms)):
    resultats[noms[i]] = scores[i]
print(resultats)

resultats['moyenne'] = round(sum(scores)/len(scores), 1)
print(resultats)

for k, v in resultats.items():
    if v > resultats['moyenne']:
        print(k, v)
````

## Partie 3 — Fonctions imbriquées

### 3.1 — Décomposer la logique

Écris deux fonctions :
- `est_valide(score)` : retourne `True` si le score est entre 0 et 100 inclus, `False` sinon
- `classifier(score)` : **appelle `est_valide()`** en premier — si le score n'est pas valide, retourne `"Score invalide"`. Sinon retourne `"Reçu"` si >= 60, `"Recalé"` sinon.

Teste avec : `85`, `42`, `-5`, `110`, `60`


````python
def est_valide(score):
    if 0 <= score <= 100:
        return True
    else:
        return False

def classifier(score):
    if est_valide(score) != True:
        print(f"{score} is: Score invalide")
    elif score >= 60:
        print(f"{score} is: Reçu")
    else:
        print(f"{score} is: Recalé")

print("Test with different score:\n", )
scores = [85, 42, -5, 110, 60]
for scorex in scores:
    classifier(score=scorex)
````

### 3.2 — Pipeline de traitement

Écris 3 fonctions courtes qui s'enchaînent :
- `extraire_nombres(texte)` : prend une string comme `"ventes: 120, retours: 15, stock: 340"` et retourne une liste des nombres entiers trouvés (utilise `.split()` et `.isdigit()`)
- `filtrer_grands(nombres, seuil=100)` : retourne uniquement les nombres >= seuil
- `rapport(texte, seuil=100)` : appelle les deux fonctions précédentes et retourne la string `"Valeurs >= 100 : [120, 340]"` (avec les vrais nombres filtrés)

Teste avec `"ventes: 120, retours: 15, stock: 340, commandes: 8, expeditions: 205"`


````python
def extraire_nombres(texte):
    nombres = []
    for morceau in texte.split():
        morceau_clean = morceau.strip(",")
        if morceau_clean.isdigit():
            nombres.append(int(morceau_clean))
    return nombres

texte = "ventes: 120, retours: 15, stock: 340, commandes: 8, expeditions: 205"
extraire_nombres(texte)
````

## Partie 4 — Boucle `while`

### 4.1 — Compteur simple

Sans utiliser `for` ni `range` :
1. Affiche tous les multiples de 7 entre 1 et 50 inclus avec une boucle `while`
2. Compte combien il y en a et affiche le résultat


````python
nombre = 7
compteur = 0

while nombre <= 50:
    print(nombre)
    compteur = compteur + 1
    nombre = nombre + 7

print("Nombre de multiples :", compteur)
````

### 4.2 — Traitement d'une file

Tu as une liste de tâches à traiter :
```python
taches = ["analyser_données", "entraîner_modèle", "évaluer_modèle", "déployer", "monitorer"]
```

Utilise une boucle `while` pour :
1. Traiter (afficher) chaque tâche une par une en la retirant de la liste avec `.pop(0)`
2. Afficher `"Tâche en cours : analyser_données"` à chaque itération
3. Quand la liste est vide, afficher `"Toutes les tâches sont terminées."`

## 🎯 Bonus — Tout combiner

Tu reçois un fichier de logs sous forme de string brute :
```python
logs_bruts = """
  alice | login | 2024-01-15
  bob | error | 2024-01-15
  alice | logout | 2024-01-15
  clara | login | 2024-01-16
  bob | login | 2024-01-16
  alice | login | 2024-01-16
  clara | error | 2024-01-16
  bob | error | 2024-01-16
"""
```

1. Écris une fonction `parser_logs(logs_bruts)` qui retourne une liste de dicts avec les clés `"utilisateur"`, `"action"`, `"date"` (chaque ligne = un dict)
2. Écris une fonction `compter_actions(logs)` qui retourne un dict `{utilisateur: {action: count}}`. Ex : `{"alice": {"login": 2, "logout": 1}, ...}`
3. Affiche pour chaque utilisateur son action la plus fréquente

> 💡 Pour la question 3 : pour chaque utilisateur, itère sur son sous-dict et trouve la clé avec la valeur max (`max(sous_dict, key=lambda k: sous_dict[k])`)


````python
def parser_logs(logs_bruts):
    logs = []

    for ligne in logs_bruts.strip().split("\n"):
        utilisateur, action, date = ligne.split("|")

        logs.append({
            "utilisateur": utilisateur.strip(),
            "action": action.strip(),
            "date": date.strip()
        })

    return logs


def compter_actions(logs):
    compteur = {}

    for log in logs:
        utilisateur = log["utilisateur"]
        action = log["action"]

        if utilisateur not in compteur:
            compteur[utilisateur] = {}

        if action not in compteur[utilisateur]:
            compteur[utilisateur][action] = 0

        compteur[utilisateur][action] += 1

    return compteur


# 1. Parser les logs
logs = parser_logs(logs_bruts)

print(logs)


# 2. Compter les actions
stats = compter_actions(logs)

print(stats)


# 3. Afficher l'action la plus fréquente par utilisateur
for utilisateur, actions in stats.items():
    action_max = max(actions, key=actions.get)

    print(f"{utilisateur} : action la plus fréquente = {action_max}")
````
