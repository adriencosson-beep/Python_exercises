## 💡 Tips

- `pd.DataFrame(liste_de_dicts)` — crée un DataFrame depuis une liste de dicts
- `.head(n)` affiche les n premières lignes (5 par défaut)
- `.info()` → types des colonnes + valeurs nulles / `.describe()` → stats descriptives
- `.shape` → `(nb_lignes, nb_colonnes)` — c'est un attribut, pas une méthode (pas de `()`)
- Sélection d'une colonne : `df["colonne"]` → Series / `df[["col1", "col2"]]` → DataFrame
- Filtre : `df[df["colonne"] > valeur]` — la condition va **entre crochets**
- Plusieurs conditions : `(cond1) & (cond2)` avec des parenthèses autour de chaque condition
- Nouvelle colonne : `df["nouvelle_col"] = df["col1"] * df["col2"]`
- `.groupby("col").agg({"col2": "sum", "col3": "mean"})` — agrège plusieurs colonnes en une fois
- `.sort_values("col", ascending=False)` — trie par ordre décroissant

## Partie 4 — GroupBy & Agrégations

### 4.1 — GroupBy simple

1. Calcule le **chiffre d'affaires total par vendeur** (`.groupby().sum()`)
2. Calcule la **quantité moyenne vendue par catégorie** (`.groupby().mean()`)
3. Calcule le **nombre de ventes par région** (`.groupby().count()`)


```python
# Ton code ici
ca_vendeur = df.groupby('vendeur')['chiffre_affaires'].sum()
display(ca_vendeur)
```

```python
mean_categorie_df = df.groupby('categorie')['quantite'].mean()
display(mean_categorie_df)
```

```python
vente_region = df.groupby('region').count()
display(vente_region)
```

### 4.2 — .agg() — plusieurs métriques en une fois

En une seule opération `.groupby().agg()`, calcule **par vendeur** :
- La **somme** du `chiffre_affaires`
- La **moyenne** du `prix_unitaire`
- Le **max** de la `quantite`

Stocke le résultat dans `stats_vendeurs` et affiche-le.

> 💡 `.agg({"col1": "sum", "col2": "mean", "col3": "max"})`


```python
df_sorted = df.groupby('vendeur').agg(
    sum_ca=('chiffre_affaires', 'sum'),
    mean_unit_price=('prix_unitaire', 'mean'),
    max_qty=('quantite', 'max')
)
print(df_sorted)
```

### 4.3 — GroupBy multi-niveaux

1. Calcule le **CA total par vendeur ET par mois** (groupby sur deux colonnes)
2. Calcule par **catégorie** : la somme du CA, la quantité totale, et le prix unitaire moyen — avec `.agg()`

> 💡 Pour grouper sur deux colonnes : `df.groupby(["col1", "col2"])`


```python
ca_total_v_m = df.groupby(['vendeur', 'mois']).agg(
    CA_total=('chiffre_affaires', 'sum')
)
print(ca_total_v_m)
```

```python
sort_categorie = df.groupby('categorie').agg(
    CA=('chiffre_affaires', 'sum'),
    qty_total=('quantite', 'sum'),
    prix_unitaire_moyen=('prix_unitaire', 'mean')
)
print(sort_categorie)
```
