## 💡 Tips (à lire avant de commencer)

- `type(x)` retourne le type d'une variable — utile pour débugger
- Les **list comprehensions** `[expr for x in iterable if condition]` sont très attendues en entretien
- Un **dictionnaire** se parcourt avec `.items()`, `.keys()`, `.values()`
- Une **fonction** se définit avec `def nom(param):` et retourne une valeur avec `return`
- `range(start, stop, step)` génère une séquence d'entiers (stop **exclu**)
- Pour tester plusieurs cas dans un `if` : `elif` (pas `else if`)
- `str.split()`, `str.strip()`, `str.lower()` sont tes amis pour manipuler les chaînes

3. Crée `resultats` : liste de chaînes `"Reçu"` si score >= 60, sinon `"Recalé"`

Affiche les 3 listes.


````python
# Ton code ici
score_bonus = [score + 5 for score in scores]
print("score_bonus = ", score_bonus)

scores_eleves = [score for score in scores if score > 75]
print("score_eleves = ", scores_eleves)

resultats = [str("Reçu") if score >= 60 else "Recalé" for score in scores]
print("resultats = ", resultats)
````

### 4.1 — if / elif / else

Écris un bloc conditionnel qui, à partir d'une variable `temperature` (en °C), affiche :
- `"Froid"` si < 10
- `"Agréable"` si entre 10 et 25 (inclus)
- `"Chaud"` si > 25

Teste avec `temperature = 8`, puis `temperature = 20`, puis `temperature = 32`.


````python
# Teste avec les 3 valeurs (tu peux faire 3 cellules ou une boucle sur une liste)
temperature_1 = 8
temperature_2 = 20
temperature_3 = 32

# My code
def temp_check(temperature: int):
    if temperature < 10:
        return "Froid"
    elif 10 <= temperature <= 25:
        return "Agréable"
    else:
        return "Chaud"

print("temp_1:", temp_check(temperature_1))
print("temp_2:", temp_check(temperature_2))
print("temp_3:", temp_check(temperature_3))
````

### 5.2 — Fonction avec valeur par défaut

Écris une fonction `salutation(prenom, langue="fr")` qui :
- Si `langue == "fr"` → retourne `"Bonjour, {prenom} !"`
- Si `langue == "en"` → retourne `"Hello, {prenom}!"`
- Sinon → retourne `"Hi, {prenom}!"`

Teste :
1. `salutation("Alice")` (langue par défaut)
2. `salutation("Bob", "en")`
3. `salutation("Carlos", "es")`


````python
def salutation(prenom: str, langue: str ="fr"):
    if langue == "fr":
        return f"Bonjour, {prenom}!"
    elif langue == "en":
        return f"Hello, {prenom}!"
    else:
        return f"Hi, {prenom}!"

print(salutation("Alice"))
print(salutation("Bob", "en"))
print(salutation("Carlos", "es"))
````

### 5.3 — Fonction + list comprehension (combinaison)

Écris une fonction `filtrer_scores(scores, seuil=60)` qui :
- Prend une liste de scores et un seuil
- Retourne **uniquement les scores >= seuil** (avec une list comprehension)

Teste avec `scores = [45, 72, 58, 88, 61, 39, 95]` et le seuil par défaut, puis avec `seuil=80`.


````python
def filter_scores(scores, seuil=60):
    filtered_scores = [score for score in scores if score >= seuil]
    return filtered_scores

print(filter_scores(scores=[45, 72, 58, 88, 61, 39, 95]))
print(filter_scores(scores=[45, 72, 58, 88, 61, 39, 95], seuil=80))
````

### Bonus — Tout combiner

Tu as la liste de dictionnaires suivante représentant des candidats :

```python
candidats = [
    {"nom": "Alice", "score": 82, "experience": 3},
    {"nom": "Bob", "score": 55, "experience": 1},
    {"nom": "Clara", "score": 91, "experience": 5},
    {"nom": "David", "score": 68, "experience": 2},
    {"nom": "Eva", "score": 74, "experience": 4},
]
```

1. Affiche le nom et le score de chaque candidat (boucle sur la liste de dicts)
2. Crée une liste `candidats_retenus` : uniquement ceux avec un score >= 70
3. Affiche combien de candidats sont retenus
4. Écris une fonction `score_moyen(liste_candidats)` qui retourne la moyenne des scores. Teste-la sur `candidats` et sur `candidats_retenus`.


````python
candidats = [
    {"nom": "Alice", "score": 82, "experience": 3},
    {"nom": "Bob", "score": 55, "experience": 1},
    {"nom": "Clara", "score": 91, "experience": 5},
    {"nom": "David", "score": 68, "experience": 2},
    {"nom": "Eva", "score": 74, "experience": 4},
]

print("---- step 1 ----")
for candidat in candidats:
    print(f"Le candidat {candidat["nom"]} a obtenu un score de {candidat["score"]}")

print("\n---- Step 2 ----")
candidats_retenus = []
for candidat in candidats:
    if candidat['score'] >= 70:
        candidats_retenus.append(candidat)
print(candidats_retenus)
print("Nombre de candidats retenus:", len(candidats_retenus))

print("\n---- Step 3 ----")
def score_moyen(liste_candidats):
    mean_scores = 0
    for candidat in liste_candidats:
        mean_scores += candidat['score']
    return mean_scores/len(liste_candidats)

print(f"\nScore moyen de tous les candidats : {score_moyen(candidats):.2f}")
print(f"Score moyen des candidats retenus : {score_moyen(candidats_retenus):.2f}")
````
